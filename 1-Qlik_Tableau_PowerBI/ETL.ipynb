```python
import pandas as pd
```

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
df1 = df.dropna()
df1 = df1.drop('Rating', axis=1)
```

```python
df1['NA_Sales'] = df1['NA_Sales'].apply(lambda x: x*1000000)
df1 = df1.astype({'NA_Sales':'int'})
df1['EU_Sales'] = df1['EU_Sales'].apply(lambda x: x*1000000)
df1 = df1.astype({'EU_Sales':'int'})
df1['JP_Sales'] = df1['JP_Sales'].apply(lambda x: x*1000000)
df1 = df1.astype({'JP_Sales':'int'})
df1['Other_Sales'] = df1['Other_Sales'].apply(lambda x: x*1000000)
df1 = df1.astype({'Other_Sales':'int'})
df1['Global_Sales'] = df1['Global_Sales'].apply(lambda x: x*1000000)
df1 = df1.astype({'Global_Sales':'int'})

df1 = df1.astype({'Year_of_Release':'int'})
df1 = df1.astype({'Critic_Count':'int'})
df1 = df1.astype({'User_Count':'int'})
df1['Critic_Score'] = df1['Critic_Score'].apply(lambda x: x/10)
```

```python
df1.to_csv('Video_Games_Sales_ETL.csv', sep=';',index=False)
```
